```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
from pprint import pprint

# Import API key
from api_keys import g_key
```

### Store Part I results into DataFrame
* Load the csv exported in Part I to a DataFrame

```python
file_path = os.path.join("output_data", "cities.csv")
weather_data = pd.read_csv(file_path)
weather_data.set_index(["City_ID"], inplace=True)
weather_data.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
print(g_key)
```

```python
# access maps with unique API key
gmaps.configure(api_key=g_key)
# create dataframe with lat, lng and Humidity
locations = weather_data[["Lat","Lng"]]
humidity = weather_data["Humidity"]
```

```python
# create intial gmap figure
fig = gmaps.figure()
# add heatmap layer
heatmap_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=0.9)
# add heatmap layer to figure
fig.add_layer(heatmap_layer)
# Display the figure
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
ideal_weather = weather_data[(weather_data["Humidity"] > 70) & (weather_data["Humidity"] <= 75)]
ideal_weather = ideal_weather.dropna()
ideal_weather
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# store ideal_weather dataframe as hotel_df
hotel_df = ideal_weather.astype(str)
# add empty Hotel Name column to dataframe
hotel_df["Hotel Name"] = ""

# set base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json?"

# parameter variables hotels within 5000 meters
for index,row in hotel_df.iterrows():
    try:
        location = row[str("Lat")] + "," + row[str("Lng")]
        request_url = base_url + "location=" + location + "&radius=5000" + "&type=lodging" + "&key=" + g_key
        response = requests.get(request_url).json()
        hotel_name = response["results"][0]["name"]
        row["Hotel Name"] = hotel_name
    except IndexError:
        row["Hotel Name"] = "**No hotel within 5000 meters**"
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations = hotel_df[["Lat", "Lng"]]
```

```python
# Add marker layer ontop of heat map
locations = locations.astype(float)
markers = gmaps.marker_layer(locations)
fig.add_layer(markers)

# Display figure
fig
```
